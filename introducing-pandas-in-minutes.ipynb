```python
# import imp packages
import pandas as pd
import numpy as np
```

```python
# create a pandas series
my_series = pd.Series([1, 3, 5, 6, 8])
my_series
```

```python
# create a time series: input the start date in format yyyymmdd
dates = pd.date_range('20170801', periods=6)
dates
```

```python
# create a matrix with 6 rows and 4 columns
# une numpy to generate random number
matr = np.random.rand(6, 4)

# use the matrix to generate a dataframe
sample_df = pd.DataFrame(matr,
                  index=dates,
                 columns=list(('A', 'B', 'C', 'D')))
sample_df
```

```python
# check data types for each column
sample_df.dtypes
```

```python
# view index of the dataframe
sample_df.index
```

```python
# display first 5 rows of data
sample_df.head()
```

```python
# display last 5 rows of data
sample_df.tail()
```

```python
# display first n rows of data
n = 3
sample_df.head(n)
```

```python
# get column names of the dataframe -> create a list out of it
col_name = list(sample_df.columns)
col_name
```

```python
# get values from each column of the dataframe
all_val = sample_df.values
print(all_val)
```

```python
# compute basic statistics measurement like mean, std, variance and etc.
sample_df.describe()
```

```python
# compute basic stats for some specific type of column
# specify the type of column you want
# our case everything is float64
sample_df.describe(include=["float64"])
```

```python
# a simple top view of the dataframe
sample_df
```

```python
# transpose the dataframe: you can use the same to do with a numpy array
sample_df.T
```

```python
# sort index by an axis
# axis = 1: row
# axis = 0: column
# sorting column wise
sample_df.sort_index(axis=0,
              ascending=False)
```

```python
# sort index row wise
sample_df.sort_index(axis=1,
              ascending=False)
```

```python
# based on the values in col A sort the entire dataframe
# result in sorted col A
sample_df.sort_values(by='A',
               ascending=True)
```

```python
# slice dataframe
# use .iloc for index -> row wise
# use .loc for names -> column wise
print(sample_df.iloc[0:3]) # first 3 rows: 0, 1, 2
print(sample_df.iloc[1:4]) # rows from 1 to 3
```

```python
# slice the dataframe using names with .loc
print(sample_df.loc[:, ['A', 'C']]) # all rows and columns A and C
print(sample_df.loc[:, ['B', 'A']]) # all rows and columns B and A
```

```python
# .loc can also be used to select specific row and column
sample_df.loc['20170801' : '20170803', ['A', 'C', 'D']]
```

```python
# use filter to select data
# select rows if data in A is greater than 0.5
print(sample_df.loc[sample_df.A > 0.5])

# select rows if data in C is greater than data in D
print(sample_df.loc[sample_df.C > sample_df.D])
```

```python
# generate true copy of a dataframe - changes to a copy is not reflected onto other
new_df = sample_df.copy()
new_df
```

```python
# adding a new column to the dataframe
new_df['E'] = ['one', 'two', 'three', 'four', 'five', 'six']
new_df
```

```python
# chanegs in new_df is not reflected in the original dataframe df
sample_df
```

```python
# filter data use select operation on columns
# select rows if column E has 'two' or 'three'
print(new_df.loc[new_df['E'].isin(['two', 'three'])])

# select rows if column E has 'four'
print(new_df.loc[new_df["E"].isin(["four"])])
```

```python
# add another column with NaN values
new_df['F'] = [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
new_df
```

```python
# change the values of column F to 1 for the selected rows based on index
new_df.loc["20170801":"20170803", "F"] = 1
new_df
```

```python
# adding a new column
new_df['G'] = [np.nan, 1.0, 2.0, 3.0, 2.0, 1.0]
new_df
```

```python
# drop the entire row with NaN value in any column
new_df.dropna(how='any')
```

```python
# drop rows with all column values as NaN
new_df.dropna(how='all')
```

```python
# fill the NaN values using fillna() function
new_df.fillna(value=10) # pass value to be filled
```

```python
# get the boolean mask for NaN of the dataframe
# returns true if Not a number else true
pd.isnull(new_df)
```

```python
# create a new pandas dataframe with 10 rows and 4 columns
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# slice the dataframe into into 3 parts
# first dataframe 0-2 row
# second dataframe 3-6 row
# third dataframe 7-9
pieces = [df[:3], df[3:7], df[7:]]
```

```python
pieces[0]
```

```python
pieces[1]
```

```python
# join the pieces together to get the original dataframe
pd.concat(pieces) # pass a list to concat all elements together
```
