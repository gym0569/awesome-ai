## Scatter Plot

```python
# import numpy for sample data generation
import numpy as np
```

```python
# get two arrays of numbers of length 10 each
x = np.arange(10)
y = np.arange(10)

y = y*2
i = 0
while i < 10:
    print("(", x[i], y[i], "), ")
    i += 1

# plot all those points using a scatter plot
plt.figure()
plt.scatter(x, y, s = 100)
plt.show()
```

```python
# creating two numpy arrays
new_x = np.array([2, 4, 7, 8])
new_y = np.array([1, 8, 3, 0])

# let us plot them using scatter plot
plt.figure()
plt.scatter(new_x, new_y, s=50, color="orange")
```

```python
# plotting a new scattter plot using an array f length 10
# first array is from 0 to 9
# second array is from 10 to 19
# s here is the size of the points

plt.figure() # new plotting canvas

plt.scatter(np.arange(0, 11), np.arange(10, 21), s = 100, color = 'orange')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis([0, 10, 10, 20])
plt.show()
```

```python
# update the previous canvas
plt.scatter(np.arange(1, 10, 2), np.arange(10, 20, 2), s = 300, color = 'green')
plt.show()
```

```python
plt.legend(['Plot 1', 'Plot 2'])
```

```python
plt.figure() # new plotting canvas

plt.scatter(np.array([2, 4, 6, 8]), np.array([2, 4, 6, 8]), s=100, color="skyblue")
plt.scatter(np.array([1, 5, 7, 1]), np.array([6, 4, 4, 3]), s=300, color="yellow")

plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(["Plot1", "Plot2"])

plt.show()
```

## Line Plots

```python
plt.figure() # initiate new canvas

y1 = np.arange(1, 10)
print("y1 =", y1)

plt.plot(y1, '-o') # only y axis values no x-axis values
plt.legend(["Line1"])
plt.axis([0, 10, 0, 10])
plt.show()
```

```python
plt.figure() # initiate new canvas

y1 = np.arange(1, 10)
y2 = np.array([1, 3, 5, 7, 9, 11, 13, 15, 17])

print("y1 =", y1)
print("y2 =", y2)

plt.plot(y1, "-o", y2, "-o") # only y axis values no x-axis values
plt.legend(["Line1", "Line2"])
plt.axis([0, 10, 0, 20]) # set axis

# let us now fill the area between these two lines
plt.gca().fill_between(range(len(y1)), y1, y2, facecolor="blue", alpha=0.70)

# let us fill only first half of it
plt.gca().fill_between(range(5), y1[0:5], y2[0:5], facecolor="orange", alpha=0.3)

# add the plotting details
plt.legend(["Plot Y1", "Plot Y2", "FullCover1", "cover2"])
plt.show()
```

## Bar Charts

```python
plt.figure() # new canvas

x = range(1, 10)
y = range(10, 19)
plt.bar(x, y, width = 0.3)
plt.show()
```

```python
# generate a new data
newx = []
for i in x:
    newx.append(i + 0.3)

# plot this data as bar graph
plt.bar(newx, y, width = 0.3, color = 'maroon')
```

```python
plt.axis([0, 10, 0, 50]) # set axis
plt.legend(['Y1', 'Y2']) # set plot details
plt.show()
```

```python
plt.legend(['Linear Y1', 'Quad Y2'])
plt.axis([0, 20, 0, 10])
```

```python
plt.figure() # initiate a new canvas

# subplotting: multiple plots in one canvas
plt.subplot(1, 2, 1) # 1 row and 2 col and start with first col
plt.plot(np.arange(10), '-o')
plt.legend(["Plot1"])

plt.subplot(1, 2, 2) # now drwa on the second column
plt.plot(np.arange(5), '-o')
plt.legend(["Plot2"])

plt.show()
```

```python
# define different subplot altogether
# share the x axis together and only change the y axis

fig, ((x1, x2), (x3, x4)) = plt.subplots(2, 2, sharex = True) # 4 plots: 2 rows and 2 columns
ax = [x1, x2, x3, x4]

for n in range(len(ax)):
    # generate some data and plot
    sample_s = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_s)
    # plot the data
    ax[n].hist(sample, bins = 100)
    # name the graphs
    ax[n].set_title('n = {}'.format(sample_s))
```

```python
plt.figure() # new canvas

# get some random data
y = np.random.normal(size = 10000)
x = np.random.random(size = 10000)

plt.scatter(x, y)
plt.show()
```

# Heatmaps

```python
plt.figure() # new canvas

# get somee random data: any normal or anything, make sure size is same
x = np.random.random(size = 10000)
y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000)

# plot a 2d histogram: basically heatmap: color changes based on the value
plt.hist2d(x, y, bins = 100)
plt.colorbar() # add the color bar on the side
plt.show()
```

# Word Cloud

```python
# data to work with
text_blob = """Computer software (often called just software) is a set of instructions and associated documentation that 
tells a computer what to do or how to perform a task or it can mean all the software on a computer, 
including the applications and the operating system."""
```

```python
# import imp packages
from wordcloud import WordCloud, STOPWORDS
```

```python
my_word_cloud = WordCloud().generate(text_blob)
```

```python
plt.figure()
plt.imshow(my_word_cloud, interpolation="bilinear")
plt.axis("off")
```

```python
# another way of doing the same thing
wordcloud = WordCloud(max_font_size=40).generate(text_blob)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS

# read the whole text from the file
text = open("datasets/alice.txt").read()

# open the mask image file and save it as a numpy array
# the mask image must be gray scale
alice_mask = np.array(Image.open("datasets/images/alice.png"))


stop_words = set(STOPWORDS)

wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stop_words)

# generate word cloud
wc.generate(text)

# plot the wordcloud
plt.figure()
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

```python
# read the whole text from the file
text = open("datasets/batman.txt").read()

# open the mask image file and save it as a numpy array
# the mask image should be only black and white to work properly
alice_mask = np.array(Image.open("datasets/images/b1.jpeg"))

wc = WordCloud(background_color="black", max_words=20000, mask=alice_mask, max_font_size=20)

# generate word cloud
wc.generate(text)

# plot the wordcloud
plt.figure()
plt.imshow(wc, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
```
