```python
# load the packages
import pandas as pd
import numpy as np
```

```python
# load data
data = pd.read_csv('price-dataset.csv', header=0)
data.head()
```

```python
# set index
data = data.set_index("Id")
data.head()
```

```python
# get shape of the data
data.shape
```

```python
# get basic stats about the numeric features
data.describe()
```

```python
# get feature list
features = list(data.columns)
```

```python
# get numeric features
numeric_features = [x for x in features if data[x].dtype != "object"]

# get categorical features
categorical_features = [x for x in features if data[x].dtypes == "object"]

print(numeric_features, categorical_features)
```

```python
# check for null or missing values
data.isnull().sum()
```

```python
# check for the percentage of null values and if more than 50% drop the feature
# else fill the missing values with mean value
for x in numeric_features:
    null_values = data[x].isnull().sum()
    per_null = null_values / data.shape[0] * 100
    if per_null > 50.0:
        del data[x]
        print(x, ": feature removed due to higher percentage of missing values!")
    else:
        data[x].fillna(np.mean(data[x]), inplace=True)
        print(x, ": missing values filled!")
```

```python
# check for the percentage of null values and if more than 50% drop the feature
# else fill the missing values with highest frequency label
removed_categorical_feature_list = list()
for x in categorical_features:
    null_values = data[x].isnull().sum()
    per_null = null_values / data.shape[0] * 100
    if per_null > 50.0:
        del data[x]
        removed_categorical_feature_list.append(x)
        print(x, ": feature removed due to high percentage of missing values!")
    else:
        data[x].fillna(data[x].value_counts().idxmax(), inplace=True)
        print(x, ": missing values filled!")
```

```python
# now lets check for null values again
data.isnull().sum()
```

```python
data.head()
```

```python
# let us normalize numeric features
for x in numeric_features:
    data[x] = data[x].apply(lambda s: (s - min(data[x])) / (max(data[x]) - min(data[x])))
data.head()
```

```python
# update the categorical feature list
updated_categorical_feature_list = list()
for x in categorical_features:
    if x not in removed_categorical_feature_list:
        updated_categorical_feature_list.append(x)

# let us encode string labesl to numeric values for categorical features
for x in updated_categorical_feature_list:
    new_feature = x + "Code"
    data[new_feature] = data[x].astype("category").cat.codes
    # create mapping
    mappings = list()
    for i, row in data[[x, new_feature]].iterrows():
        tmp = str(row[x]) + "->" + str(row[new_feature])
        if tmp not in mappings:
            mappings.append(tmp)
    # save the mapping for further use
    file = open(x+"_"+new_feature+"_Mapping.txt", mode="w")
    for m in mappings:
        file.writelines(str(m)+"\n")
    file.close()
```

```python
data.head()
```

```python
# remove redundant categorical features
for x in updated_categorical_feature_list:
    del data[x]
data.head()
```

```python
# save the preprocessed data for modelling
data.to_csv("train.csv")
```
