## Scatter Plot

```python
import numpy as np
```

```python
x = np.arange(10)
y = np.arange(10)
```

```python
plt.figure()
plt.scatter(x, y, s = 100)
```

```python
new_x = np.array([2, 4, 7, 8])
new_y = np.array([1, 8, 3, 0])
```

```python
plt.scatter(new_x, new_y, s = 100, color = 'red')
```

```python
plt.figure() #new plot
plt.scatter(np.arange(0, 10), np.arange(10, 20), s = 100, color = 'orange')
```

```python
plt.scatter(np.arange(1, 10, 2), np.arange(10, 20, 2), s = 300, color = 'green')
```

```python
plt.legend(['Plot 1', 'Plot 2'])
```

## Line Plots

```python
plt.figure()
y1 = np.arange(1, 10)
y2 = y1 ** 2
plt.plot(y1, '-o', y2, '-o') #only y axis values no x-axis values
```

```python
plt.axis([0, 10, 1, 100])
```

```python
plt.gca().fill_between(range(len(y1)), y1, y2, facecolor = 'blue', alpha = 0.40)
```

```python
plt.legend(['Plot Y1', 'Plot Y2', 'Cover'])
```

## Bar Charts

```python
plt.figure()
x = range(len(y1))
plt.bar(x, y1, width = 0.3)
```

```python
newx = []
for i in x:
    newx.append(i + 0.3)

plt.bar(newx, y2, width = 0.3, color = 'maroon')
```

```python
plt.axis([-1, 10, 0, 100])
```

```python
plt.legend(['Y1', 'Y2'])
```

```python
plt.figure()
x = range(len(y1))
plt.bar(x, y1, width = 0.5, color = 'r')
plt.bar(x, y2, width = 0.5, color = 'b', bottom = y1)
```

```python
plt.legend(['Linear Y1', 'Quad Y2'])
```

```python
plt.figure()
x = range(len(y1))
plt.barh(x, y1, height = 0.5, color = 'r')
plt.barh(x, y2, height = 0.5, color = 'b', left = y1)
```

```python
plt.legend(['Linear Y1', 'Quad Y2'])
```

```python
plt.axis([0, 80, 0, 9])
```
