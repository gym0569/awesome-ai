## Histogram

```python
#define different subplot altogether
#share the x axis together and only change the y axis
fig, ((x1, x2), (x3, x4)) = plt.subplots(2, 2, sharex = True)
ax = [x1, x2, x3, x4]

#generate some data and plot
for n in range(len(ax)):
    sample_s = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_s)
    ax[n].hist(sample, bins = 100)
    #name the graphs
    ax[n].set_title('n = {}'.format(sample_s))
```

```python
plt.figure()
y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000)
x = np.random.random(size = 10000)
plt.scatter(x, y)
```

```python
import matplotlib.gridspec as gspec
```

```python
plt.figure()

#three subplot
g = gspec.GridSpec(3, 3)

top_h = plt.subplot(g[0, 1:])
side_h = plt.subplot(g[1:, 0])
lower_h = plt.subplot(g[1:, 1:])
```

```python
#generate some data
x = np.random.random(size = 10000)
y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000)

top_h.hist(x, bins = 100)
lower_h.scatter(x, y)
s = side_h.hist(y, bins = 100, orientation = 'horizontal')
```
