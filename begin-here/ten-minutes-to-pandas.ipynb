Pandas primarily uses the value np.nan to represent missing data points. It is by default not inclued in computations. Re-indexing allows to change/add/delete the index on a specific axis.

```python
# add another column with NaN values
df2['F'] = [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
df2
```

```python
# change the values of column F to 1 for the selected rows
df2.loc[dates[0]:dates[2], 'F'] = 1
df2
```

```python
# adding a new column
df2['G'] = [np.nan, 1.0, 2.0, 3.0, 2.0, 1.0]
df2
```

```python
# drop the entire row with NaN value in any column
clean_data = df2.dropna(how = 'any')
clean_data
```

```python
df2
```

```python
# fill the NaN values using this function
# data type should match
df2.fillna(value = 9.9)
```

```python
# create a new pandas dataframe: 10x4
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# slice the dataframe into into 3 parts
pieces = [df[:3], df[3:7], df[7:]]
pieces
```

```python
# join the pieces together to get the original dataframe
pd.concat(pieces)
```

```python
# create new data frames using a dictionary
left = pd.DataFrame({'key': ['foo', 'foo'], 'val1': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'val2': [3, 4]})
print(left)
print(right)
```

```python
left
```

```python
right
```

```python
# merge based on key
pd.merge(left, right, on='key')
```

```python
pd.merge(right, left, on='key')
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
```
