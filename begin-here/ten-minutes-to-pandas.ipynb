```python
"""
slicing: selection method - .iloc and .loc
.iloc for index positioning
.loc for column name indexing
"""

print(df.iloc[0:3]) # first 3 rows: 0, 1, 2
print(df.iloc[1:4]) # rows from 1 to 3
```

```python
# another type of slicing using column names
print(df.loc[:, ['A', 'C']]) # all rows and columns A and C
print(df.loc[:, ['B', 'A']]) # all rows and columns B and A
```

```python
# select specific rows with specific columns based on the index: cannot use row number here
df.loc['20170801' : '20170803', ['A', 'C', 'D']]
```

```python
# use filter to select data

# select rows with data in A is greater than 0
print(df.loc[df.A > 0.5])

# select rows with data in C is greater than data in D
print(df.loc[df.C > df.D])
```

```python
# generate true copy of a dataframe - changes to a copy is not reflected onto other
df2 = df.copy()
df2
```

```python
# adding a new column to the dataframe
df2['E'] = ['one', 'two', 'three', 'four', 'five', 'six']
df2
```

```python
# chanegs in df2 is not reflected in the original dataframe df
df
```

```python
# filter data use select operation on columns

# select rows if column E has 'two' or 'three'
print(df2.loc[df2['E'].isin(['two', 'three'])])

# select rows if column E has 'four'
print(df2.loc[df2["E"].isin(["four"])])
```

```python
"""
Pandas primarily uses the value np.nan to represent missing data points.
It is by default not inclued in computations. Re-indexing allows to change/add/delete the index on a specific axis.
"""

# add another column with NaN values
df2['F'] = [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
df2
```

```python
# change the values of column F to 1 for the selected rows based on index
df2.loc["20170801":"20170803", "F"] = 1
df2
```

```python
# adding a new column
df2['G'] = [np.nan, 1.0, 2.0, 3.0, 2.0, 1.0]
df2
```

```python
# drop the entire row with NaN value in any column
print(df2.dropna(how='any'))

# drop rows with alll column values as NaN
print(df2.dropna(how='all')) # in our case no drop since no such row is present
```

```python
# fill the NaN values using fillna() function

df2.fillna(value=10)
```

```python
# get the boolean mask for NaN of the dataframe
print(df2)

print(pd.isnull(df2))
```

```python
# create a new pandas dataframe with 10 rows and 4 columns

df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# slice the dataframe into into 3 parts
pieces = [df[:3], df[3:7], df[7:]]
print(pieces[0])
print(pieces[1])
print(pieces[2])
```

```python
# join the pieces together to get the original dataframe
pd.concat(pieces) # or pd.concat(pieces[0],pieces[2],pieces[2])
```

```python
# create new data frames using a dictionary
left = pd.DataFrame(
    {
        "key": ["Fruit", "Vegetable"],
        "val1": ["Apple", "Potato"]
    })
right = pd.DataFrame(
    {
        "key": ["Fruit", "Fruit"],
        "val2": ["Banana", "Mango"]
    })
```

```python
left
```

```python
right
```

```python
# merge based on key
pd.merge(left, right, on='key')
```

```python
# another way to merge
pd.merge(right, left, on='key')
```

```python
# create a dataframe from a dictionary
df = pd.DataFrame(
    {
        "Id":[1, 2, 3, 4, 5],
        "Grade":["A", "B", "A", "C", "B"]
    })
df # as you can see, columns are arranged in alphabetical order
```
