## Preprocessing

```python
# LOAD PACKAGES
import pandas as pd
import numpy as np
```

```python
# LOAD THE DATA
data = pd.read_csv("data.csv")
data.head()
```

```python
# remove useless lsat column
data = data[[col for col in data.columns if col != "Unnamed: 32"]]
data.head()
```

```python
print("number of rows = " ,data.shape[0], "number of columns = ", data.shape[1])
print("columns = ", data.columns)
```

```python
# check for null or missing values
data.isnull().sum()
```

```python
# check data types of all the column
data.dtypes
```

```python
# set numeric codes for the diagnosis column
# get the unique values present for the column
data['diagnosis'].unique()
```

```python
# create a map
mapping = {'M': 1, 'B': 0} # 1 = tumour and 0 = no-tumour

# map the string to int
data['diagnosis'] =  data['diagnosis'].map(mapping)
data.head()
```

```python
# normalize data if needed
# we dont consider id column here
for col in data.iloc[:, 1:].columns:
    min_d = min(data[col])
    max_d = max(data[col])
    print("column = ", col, ":", min_d, max_d)
    if min_d >= -1 and max_d <= 1:
        continue
    else:
        print("normalizign data...")
        data[col] = data[col] / max(data[col])
    print()
```

```python
data.head()
```

```python
# What is the number of Patients with and without tumour in the dataset
patients = data['diagnosis'].value_counts()
print("No Tumour = ", patients[0])
print("With Tumour = ", patients[1])
```

```python
print("Percntage of Patients with Tumour = ", patients[1] / data.shape[0])
print("Percntage of Patients with No-Tumour = ", patients[0] / data.shape[0])
```

```python
# check for highly skwed data
for col in data.iloc[:, 2:].columns:
    skweness = data[col].skew()
    if skweness >= 1.0 or skweness <= -1.0:
        print(col, format("data is highly skwed", "*^30"))
    else:
        continue
```

```python
# perform log transformation for highly skwed data
for col in data.iloc[:, 2:].columns:
    skweness = data[col].skew()
    if skweness >= 1.0 or skweness <= -1.0:
        data[col] = data[col].apply(lambda x: np.log(x))
    else:
        continue
```

```python
# now as you can see no highly skwed data column
data.head()
```
