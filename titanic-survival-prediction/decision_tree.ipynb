```python
# LOAD PACKAGES
import numpy as np
import pandas as pd
```

```python
# LOAD DATA
train_file = "data/train.csv"
test_file = "data/test.csv"

train = pd.read_csv(train_file)
test = pd.read_csv(test_file)
```

```python
train.head(5) # VIEW THE DATA
```

```python
test.head(5)
```

```python
# REMOVE NAME COLUMN SINCE IT IS OF VERY LITTLE USE IN SURVIVAL PREDICTION
# ALSO REMOVE TICKET NUMBER: GENERALLY IT WILL BE A RANDOM NUMBER
# REMOVE CABIN SINCE IT IS RELATED TO CLASS AND EMBARKED INDIRECTLY
train = train.loc[:, [col for col in train.columns if col not in ["Name", "Ticket", "Cabin"]]]

# DONT FORGET TP REMOVE IT FROM TEST DATASET ALSO
test = test.loc[:, [col for col in test.columns if col not in ["Name", "Ticket", "Cabin"]]]
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
# LET US ENCODE EMBARKED COLUMN
train["EmbarkedCode"] = train.Embarked.astype("category").cat.codes
# SAME FOR TEST DATA
test["EmbarkedCode"] = test.Embarked.astype("category").cat.codes

# ENCODE SEX COLUMN AS WELL
train["SexCode"] = train.Sex.astype("category").cat.codes
# SAME FOR TEST DATA
test["SexCode"] = test.Sex.astype("category").cat.codes
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
# NOW LET US NORMALIZE THE FARE COLUMN
train.Fare = train.Fare / max(train.Fare)
# SAME FOR TEST
test.Fare = test.Fare / max(test.Fare)
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
# SELECT COLUMN TO USE AS FEATURES NOW
train = train.loc[:, [col for col in train.columns if col not in ["Sex", "Embarked"]]]
# DO SAME FOR TEST DATA
test = test.loc[:, [col for col in test.columns if col not in ["Sex", "Embarked"]]]
```

```python
# CHECK FOR MISSING VALUES
for col in train.columns:
    print(train[[col]].isnull().sum())
    print()
```

```python
# AS WE CAN SEE AGE HAS 177 MISSING VALUES
# ONE WAY IS TO REMOVE THEM BUT SINCE DATA IS SMALL WE WILL FILL IT RATHER THAN REMOVING IT

# FILLING MEAN OF THE AGE FOR THE MISSING VALUES
train.Age.fillna(train.Age.mean(), inplace=True)
```

```python
# NOW LETS AGAIN CHECK IF IT WORKED OR NOT
for col in train.columns:
    print(train[[col]].isnull().sum())
```

```python
# CHECK FOR NULL VALUES IN TEST DATASET ALSO
for col in test.columns:
    print(test[[col]].isnull().sum())
```

```python
# FILL AGE MISSING VAKUES WITH MEAN AGE AS EARLIER
test.Age.fillna(test.Age.mean(), inplace=True)

# FILL MISSING VALUES IN FARE WITH ITS MEDIUM VALUE
test.Fare.fillna(test.Fare.median(), inplace=True)
```

```python
# NOW LETS CHECK AGAIN IF IT WORKED
# NOTE: ITS ALWAYS GOOD HABBIT TO CHECK AND MOVE FURTHER EFFECT OF OPERATIONS PERFORMED
for col in test.columns:
    print(test[[col]].isnull().sum())
```

Now we are good to go!

```python
# LETS SAVE OUR CLEANED DATA SO THAT WE DONT HAVE TO CLEAN AGAIN AND AGAIN: SAVE COMPUTATION POWER
train.to_csv("data/train_cleaned.csv", index=False)
test.to_csv("data/test_cleaned.csv", index=False)
```

BUILD MODEL USING MACHINE LEARNING ALGOITHM : DECISION TREE

```python
# LOAD DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
# SPLIT DATA INTO TRAIN AND VALIDATION DATASET
x_train, x_val, y_train, y_val = train_test_split(train.iloc[:, 2:],
                                                  train.iloc[:, 1:2],
                                                  test_size = 0.3,
                                                  random_state = 49)
```

```python
print(x_train.columns, y_train.columns)
print(x_val.columns, y_val.columns)
```

```python
# MODEL A DECISION TREE CLASSIFIER
clf = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=6, random_state=123)

# FIT OUR MODEL ON THE TRAINING DATA
clf.fit(x_train, y_train)
```

```python
predictions = clf.predict(x_val)
```

```python
# NOW CHECK FOR ACCURACY
from sklearn.metrics import accuracy_score
```

```python
# NOW LETS VIEW ACCURACY OF OUR MODEL
accuracy_score(y_val, predictions)
```
