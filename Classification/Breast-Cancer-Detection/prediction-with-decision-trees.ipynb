```python
# fire up the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.DataFrame.from_csv("cancer-dataset.csv")
data.head()
```

```python
data = data.reset_index()
data.head()
```

As we can see, we have our data loaded and packages are fired up.
First phase will be data cleaning part and feature engineering. The amount of effort and time in phase depends on the quality of the data.

```python
print("Columns = ", data.columns)
```

```python
print("Dimension of the data = " ,data.shape)
```

```python
# check for missing/nan values
pd.isnull(data) # gives a boolean mask
```

```python
# find the number of null values for each column/feature
column_names = data.columns
i = 0
for x in column_names:
    i += 1
    print("Column", i, ":", data[x].isnull().sum(axis=0)) # compute number of such values for each column in the dataset
```

```python
# as the last column has all its value null, we can easily drop that feature
data = data.drop("Unnamed: 32", 1)
```

```python
data.head()
```

```python
data.dtypes
```

```python
# let us map the object values to int for diagnosis feature
print("Unique values: ", data['diagnosis'].unique())
```

```python
values = {'M': 1, 'B': 0}
# malignant = found, benign = not found
data['diagnosis'] =  data['diagnosis'].map(values)
data.head()
```

Most of the time, data cleaning takes a lot of time and effort but thankfully thats not the case here. Dataset is fairly cleana and sorted. Now as the data is clean and in proper format for any algorithm to take in, we go further with data visualization.
In the next phase we try to understand and answer some of the basic question in our mind.

```python
# before we start let us save the cleaned data for further use
data.to_csv("new-cancer-dataset.csv", index=False)
```

```python
# let us now plot the effect of each of these features on the diagnosis data points
# the function displays graph step wise
# enter "." to plot next graph and "done" to exit in mid way

def display_scatter_graph():
    column_names = data.columns
    y = data.loc[:, 'diagnosis']
    for name in column_names:
        print("Feature name = ", name)
        prompt = input()
        if prompt == ".":
            print("Graph: X =", name, "| Y = diagnosis")
            x = data.loc[:, name]
            plt.scatter(x, y, s = 75, alpha = "0.3", marker="o")
            plt.xlabel(name)
            plt.ylabel('diagnosis')
            plt.show()
        elif prompt == "done":
            return
```
