```python
# import imp packages
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
plt.style.use("seaborn")
```

```python
# get raw data
data = pd.DataFrame.from_csv("data/recipy_dataset.csv", header=0)
print(data.shape)
data.head()
```

```python
for i, row in data.iterrows():
    # print(len(row["Ingredients"]))
    if len(row["Ingredients"]) == 2:
        data.drop(i, inplace=True)
data.head()
```

```python
# get ingredients
count1 = list()
for i, row in data.iterrows():
    new_list = re.findall("[a-zA-z]+[ a-zA-z]*", row["Ingredients"])
    new_list2 = re.findall("\w+[a-zA-Z ]*", row["Ingredients"])
    count1.append(len(new_list2))
    # print(new_list, len(new_list))
    ing_list = list()
    for word in new_list:
        if len(word) > 4:
            ing_list.append(word)
    row["Ingredients"] =  ing_list
data.head()
```

```python
# get main ingredients
count2 = list()
ingredient_list = list()
for i, row in data.iterrows():
    new_list = re.findall("[a-zA-z]+[a-zA-Z]*", row["Main_Ingredients"])
    new_list2 = re.findall("\w+[a-zA-Z ]*", row["Main_Ingredients"])
    count2.append(len(new_list2))
    ing_list = list()
    for word in new_list:
        if len(word) > 3:
            ing_list.append(word)
            ingredient_list.append(word)
    row["Main_Ingredients"] = ing_list
data.head()
```

```python
fp = open("data/ingredient_list.txt", mode="w")
for x in ingredient_list:
    fp.write(x+",")
fp.close()
```

```python
# get all ingredients from east
east_ing_dict = dict()

for i, row in east_data.iterrows():
    val = str(row["Ingredients"])
    val = re.findall("\w+", val)
    for x in val:
        x = x.upper()
        if len(x) > 3:
            if x not in east_ing_dict:
                east_ing_dict[x] = 1
            else:
                east_ing_dict[x] += 1
sorted(east_ing_dict)

east_ing_list = list()

for x in east_ing_dict:
    east_ing_list.append(x)
len(east_ing_list)
```

```python
# get all ingredients from west
west_ing_dict = dict()

for i, row in west_data.iterrows():
    val = str(row["Ingredients"])
    val = re.findall("\w+", val)
    for x in val:
        x = x.upper()
        if len(x) > 3:
            if x not in west_ing_dict:
                west_ing_dict[x] = 1
            else:
                west_ing_dict[x] += 1
sorted(west_ing_dict)

west_ing_list = list()

for x in west_ing_dict:
    west_ing_list.append(x)
len(west_ing_list)
```

```python
# get all ingredients from north
north_ing_dict = dict()

for i, row in north_data.iterrows():
    val = str(row["Ingredients"])
    val = re.findall("\w+", val)
    for x in val:
        x = x.upper()
        if len(x) > 3:
            if x not in north_ing_dict:
                north_ing_dict[x] = 1
            else:
                north_ing_dict[x] += 1
sorted(north_ing_dict)

north_ing_list = list()

for x in north_ing_dict:
    north_ing_list.append(x)
len(north_ing_list)
```

```python
# get imp ingredients from south
south_ing_dict = dict()

for i, row in south_data.iterrows():
    val = str(row["Ingredients"])
    val = re.findall("\w+", val)
    for x in val:
        x = x.upper()
        if len(x) > 3:
            if x not in south_ing_dict:
                south_ing_dict[x] = 1
            else:
                south_ing_dict[x] += 1
sorted(south_ing_dict)

south_ing_list = list()

for x in south_ing_dict:
    south_ing_list.append(x)
len(south_ing_list)
```

```python
all_ingredients = east_ing_list + west_ing_list + north_ing_list + south_ing_list
print("All ingredients =", len(all_ingredients))

# get unique ingredients
ingredients = list()
for x in all_ingredients:
    if x not in ingredients and x.lower() not in stop_words2 and x.lower() not in stop_words:
        ingredients.append(x)
print("All usefull ingredients =", len(ingredients))
```

```python
# we now add it together to get our new dataframe
final_data = pd.concat([north_data, south_data, east_data, west_data])
final_data.shape
```

```python
# add ingredients as features to the new dataframe
cnt = 0
for ing in ingredients:
    cnt += 1
    main_list = list()
    for i, row in final_data.iterrows():
        # iterate over all the rows
        ingredient_string = str(row["Ingredients"])
        ingredient_list = re.findall("\w+", ingredient_string)
        ingredient_list = [x.upper() for x in ingredient_list]
        if ing in ingredient_list:
            main_list.append(1)
        else:
            main_list.append(0)
    final_data[ing] = main_list
    #print(cnt, ing, "added to the dataset")
    print(".", end="")
print("Done!")
```

## Select important features

```python
# import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier() # initiate
```

```python
# Feature Extraction with RFE
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(model, n_features_to_select=40)
```

```python
rfe
```

```python
fitted_model = rfe.fit(trainX, trainY) # fit the model and find the optimum features
```

```python
true_features = fitted_model.support_

selected_feature_list = list()

for i in range(len(true_features)):
    if true_features[i] == True:
        selected_feature_list.append(features[i])
selected_feature_list
```
