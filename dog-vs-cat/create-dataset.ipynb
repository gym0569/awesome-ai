```python
read_path = "data/dog/"
```

```python
# name convention = name + label + number
# cat label = 0
# dog label = 1
# trasnfer images into train and test directories
for i in range(num_dog_images):
    # read the image
    image = cv2.imread(read_path + str(i) + ".jpg")
    # now write the image to another location
    if i < 2500:
        new_file_name = test_path + "image_1_" + str(i) + ".jpg"
        print("writing file:", new_file_name)
        cv2.imwrite(new_file_name, image)
    else:
        j = i - 2500
        new_file_name = train_path + "image_1_" + str(j) + ".jpg"
        print("writing file:", new_file_name)
        cv2.imwrite(new_file_name, image)
```
